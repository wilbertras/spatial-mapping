```python
target_kid = 1
filtered_data = df_results[(df_results['KID'] == target_kid)]
powers_KID = np.unique(filtered_data['Power'].values)

T_target = 20*1e-3

Qi = []

for P_target in powers_KID: 

    dfiltered_data = filtered_data[(filtered_data['Power'] == P_target)]
    T = dfiltered_data['Temperature'].values
    Tx = np.argmin(np.abs((T - T_target)))
    Tchosen = T[Tx]
    tfiltered_data = dfiltered_data[(dfiltered_data['Temperature'] == Tchosen)]
    Qi.append(tfiltered_data['Qi'].values)
    
```

```python
fig, ax = plt.subplots()
ax.plot(powers_KID, Qi)
ax.set_yscale('log')
```
