```python
import numpy as np
```

```python
def magnification(si, so):
    return -si/so

def distance(s, f):
    return 1/(1/f - 1/s)

def focallength(s, so):
    return 1/(1/so + 1/si)

f = 300
so = 900
si = distance(so, f)

print('%s, si=%d, so=%d, M=%.1f' % ('camera', si, so, magnification(si, so)))

f = 300
si = 650
so = distance(si, f)

print('%s, so=%d, si=%d, M=%.1f' % ('phone', so, si, magnification(si, so)))
```
