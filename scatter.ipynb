```python
import numpy as np
from numpy.random import normal
import matplotlib.pyplot as plt

%matplotlib widget
```

```python
def linear(x, a, b):
    return a*x+b

def secorder(x, a, b, c):
    return a*x**2+b*x+c

def root(x, a, b):
    return a*x**.5+b

f0s = np.linspace(4, 8, 1000)
scatter = 1e-3
a = .9
b = 0.01
dff = normal(0, scatter, f0s.shape)
```

```python
fs = dff*f0s + f0s
fig, axes = plt.subplot_mosaic('abc', figsize=(9,3), constrained_layout=True)
ax = axes['a']
ax.scatter(f0s, f0s)
ax.scatter(f0s, fs)
print(np.std((fs-f0s)/f0s))
ax = axes['b']
ax.scatter(f0s, (fs-f0s), c='tab:orange')
ax=axes['c']
dfs = fs[1:]-fs[:-1]
ax.scatter(f0s[1:], dfs)
print(np.sum(np.absolute(dfs)>0.01))
```

```python
fs = dff*f0s + linear(f0s, 0.9, 0)
fig, axes = plt.subplot_mosaic('abc', figsize=(9,3), constrained_layout=True)
ax = axes['a']
ax.scatter(f0s, f0s)
ax.scatter(f0s, fs)
print(np.std((fs-f0s)/f0s))
ax = axes['b']
ax.scatter(f0s, (fs-f0s), c='tab:orange')
ax=axes['c']
dfs = fs[1:]-fs[:-1]
ax.scatter(f0s[1:], dfs)
print(np.sum(np.absolute(dfs)>0.01))
```
